```python

# Cell 1: Creating my first time-series list
spy_prices = [450.10, 452.20, 449.50, 453.00, 455.10, 454.80, 456.20, 451.90, 458.00, 460.50]
print(spy_prices)
```

```python
market_data = {'SPY':spy_prices} #this is creating dictionary for SPY in cabinet of market data
print('looking up for SPY.....')
print(market_data['SPY'])
```

```python
print('absolute value of changes in stocks')

for i in range(1, len(spy_prices)):
  daily_drift = abs(spy_prices[i] - spy_prices[i-1])
  print(f'day {i} : ',round(daily_drift, 2))
```

```python
spy_volality = []
for i in range(1, len(spy_prices)):
  daily_drift = abs(spy_prices[i] - spy_prices[i-1])
  spy_volality.append(round(daily_drift, 2))

print(spy_volality)
```

```python
# Cell 4: Defining our reusable function
def calc_daily_diff(X):
    volatility_value = []

    for i in range(1, len(X)):
        daily_diff = abs(X[i] - X[i-1])
        volatility_value.append(round(daily_diff, 2))

    return volatility_value

# Now, we 'call' the function using our original spy_prices list
spy_volatility = calc_daily_diff(spy_prices)
print("Result from function:")
print(spy_volality)
```
