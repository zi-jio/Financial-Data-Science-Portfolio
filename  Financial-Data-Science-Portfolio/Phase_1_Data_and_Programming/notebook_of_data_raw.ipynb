```python
BUMI = yf.Ticker('BUMI.jk')

#basic info
price = BUMI.info.get('regularMarketPrice', BUMI.info.get('currentPrice'))
market_cap = BUMI.info.get('marketCap')

print(f"current price: {price}")
print(f"current market cap: {market_cap}")

#getting price history

macro_tape = BUMI.history(period='1d', interval='1m')
macro_tape.head(10)
```

```python
macro_tape.to_csv("harga saham bumi")
```

```python
import yfinance as yf
import os

portfolio_universe = ['SPY', 'GLD', 'XISB.JK', '^TNX']
folder = 'portfolio_universe_week_1'
os.makedirs(folder, exist_ok=True)

for symbol in portfolio_universe:
  pointer = yf.Ticker(symbol)
  harga = pointer.history(period='10y', interval='1d')
  clean_name = symbol.replace('^', '')
  harga.to_csv(f'{folder}/{clean_name}.csv')

  print(f'downloaded {symbol}')


#
```
